## დავალება 1

მოცემულია ცხრილი employee_satisfaction.csv, რომელიც შეიცავს თანამშრომელთა კმაყოფილების კვლევის შედეგებს.
  - ჩატვირთეთ ცხრილი Pandas DataFrame-ში.
  - გამოთვალეთ და დაბეჭდეთ კმაყოფილების ქულების საშუალო, მედიანა, მოდა, დისპერსია, სტანდარტული გადახრა, ასიმეტრია და ექსცესი.
  - ააგეთ კმაყოფილების ქულების ჰისტოგრამა და box plot. გამოიყენეთ Pandas.
  - ინტერპრეტაცია გაუკეთეთ ყველა გამოთვლილ სტატისტიკურ მაჩვენებელს.

```python
# ვაიმპორტებთ საჭირო ბიბლიოთეკებს
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode


# მონაცემების წაკითხვა employee_satisfaction.csv-დან
df=pd.read_csv("employee_satisfaction.csv")


# გამოთვალეთ და დაბეჭდეთ კმაყოფილების ქულების
# საშუალო,მედიანა, მოდა, დისპერსია, სტანდარტული გადახრა, ასიმეტრია და ექსცესი.
average=df['satisfaction_score'].mean()
median=df['satisfaction_score'].median()
mode=df['satisfaction_score'].mode()[0]
var=df['satisfaction_score'].var()
std=df['satisfaction_score'].std()
skew=skew(df['satisfaction_score'])
kurt=kurtosis(df['satisfaction_score'])

print(f"საშუალო: {average}")
print(f"მედიანა: {median}")
print(f"მოდა: {mode}")
print(f"დისპერსია: {var}")
print(f"სტანდარტული გადახრა: {std}")
print(f"ასიმეტრია: {skew}")
print(f"ექსცესი: {kurt}")

# ჰისტოგრამა და box plot
plt.figure(figsize=(12, 5))

# ჰისტოგრამა
plt.subplot(1, 2, 1)
plt.hist(df['satisfaction_score'], bins=20, color='skyblue', edgecolor='blue')
plt.title('ჰისტოგრამა - კმაყოფილების ქულები')
plt.xlabel('კმაყოფილების ქულა')
plt.ylabel('ჩამოწერა')

# box plot
plt.subplot(1, 2, 2)
plt.boxplot(df['satisfaction_score'], vert=False)
plt.title('Box Plot - კმაყოფილების ქულები')
plt.xlabel('კმაყოფილების ქულა')

plt.tight_layout()
plt.show()



```

**ინტერპრეტაცია**
- **საშუალო (Mean)** - ასახავს თანამშრომელთა კმაყოფილების საშუალო დონეს
- **მედიანა (Median)** - აჩვენებს მონაცემთა შუა წერტილს.მაგ: თუ საშუალო და მედიანა ერთმანეთს ემთხვევა ან ახლოსაა, მონაცემები თანაბრადაა განაწილებული
- **მოდა (Mode)** - ასახავს კმაყოფილების ქულას, რომელიც ყველაზე ხშირად გვხვდება მონაცემებში
- **დისპერსია (Variance)** -  დისპერსია ასახავს მონაცემების საშუალოდან "გაბნევის"დონეს
- **სტანდარტული გადახრა (Standard Deviation)** - სტანდარტული გადახრა არის საშუალოდან გადახრების სიდიდე.
- **ასიმეტრია (Skewness)** - ასიმეტრია აჩვენებს მონაცემების განაწილების სიმეტრიულობას.
- **ექსცესი (Kurtosis)** - ექსცესი განსაზღვრავს განაწილების ერთგვარ "წვეტიანობას" მონაცემებში.

**ვიზუალები**
- **ჰისტოგრამის** საშუალებით შეგვიძლია დავინახოთ, მონაცემებზე დაფუძნებით, კმაყოფილების ქულების განაწილება
- **Box Plot** საშუალებას გვაძლევს გამოვიკვლიოთ მონაცემებში არსებული უკიდურესი მნიშვნელობები და განაწილების ასიმეტრია.

## დავალება 2

მოცემულია ცხრილი salaries.csv, რომელიც შეიცავს ინფორმაციას თანამშრომელთა ხელფასების შესახებ. იპოვეთ ამოვარდნილი მნიშვნელობები z-ქულის მეთოდით.
  - გამოიყენეთ NumPy და Pandas z-ქულების გამოსათვლელად.
  - იპოვეთ გადახრები z-ქულების საფუძველზე.


```python
# ვაიმპორტებთ საჭირო ბიბლიოთეკებს
import pandas as pd
import numpy as np


# მონაცემების წაკითხვა salaries.csv-დან
df = pd.read_csv("salaries.csv")

# z-ქულების გამოთვლა
mean_salary = df['salary'].mean()  # ხელფასების საშუალო
std_salary = df['salary'].std()    # ხელფასების სტანდარტული გადახრა

# z-ქულის გამოთვლა (z = (x - mean) / std)
z_scores = (df['salary'] - mean_salary) / std_salary


# ამოვარდნილი მნიშვნელობები
outliers = df[np.abs(z_scores) > 3]

# შედეგების ბეჭდვა
print("ამოვარდნილი მნიშვნელობები z-ქულების მეთოდით:")
print(outliers)
```

## დავალება 3

მწარმოებელი ამტკიცებს, რომ მისი პროდუქტის წუნდების პროცენტული მაჩვენებელი 5%-ზე ნაკლებია. შერჩევაში ზომით 100 ერთეული აღმოჩნდა 8 წუნდებული. შეამოწმეთ მწარმოებლის მტკიცება 0.05 მნიშვნელოვნების დონეზე.
  - ჩამოაყალიბეთ ნულოვანი და ალტერნატიული ჰიპოთეზები.
  - გამოთვალეთ z-ქულა პროპორციისთვის.
  - შეამოწმეთ p-მნიშვნელობა და გააკეთეთ დასკვნა.
  - ინტერპრეტაცია გაუკეთეთ შედეგებს.



```python
import scipy.stats as stats

# მოცემული მონაცემები
n = 100                # შერჩევის ზომა
x = 8                  # წუნდებულების რაოდენობა
p0 = 0.05              # საცდელი პროპორცია
p_hat = x / n          # შერჩევის პროპორცია


print("ჰიპოთეზების ტესტირება პროპორციისთვის")
print("-" * 50)
print("H₀ (ნულოვანი ჰიპოთეზა): წუნდების პროცენტი მეტია ან ტოლი 5%-ზე  - > 𝑝 ≥ 0.05")
print("Hₐ (ალტერნატიული ჰიპოთეზა): წუნდების პროცენტი 5%-ზე ნაკლებია - > 𝑝 < 0.05")

# z-ქულის გამოთვლა
z = (p_hat - p0) / ((p0 * (1 - p0) / n) ** 0.5)

# p-მნიშვნელობის გამოთვლა
p_value = stats.norm.cdf(z)

print("-" * 50)

print(f"z-ქულა: {z:.4f}")
print(f"p-მნიშვნელობა: {p_value:.4f}")

if p_value < 0.05:
    print("დადასტურდა, რომ წუნდების პროცენტი 5%-ზე ნაკლები არ არის. 𝑝 ≥ 0.05")
else:
    print("არ დადასტურდა, რომ წუნდების პროცენტი 5%-ზე ნაკლები არ არის. 𝑝 < 0.05")
```
